#### 2.3.3. while loop

```python
print('Welcome to the GPA calculator.')
print('Please enter all your letter grades, one per line.')
print('Enter a blank line to designate the end.')

# map from letter grade to point value: "dictionary object"
points = {'A+':4.5, 'A':4.0, 'B+':3.5, 'B':3.0, 
          'C+':2.5, 'C':2.0, 'D+':1.5,'D':1.0, 'F':0.0}

# initialize variables
num_courses = 0
total_points = 0
done = False

# while loop
while not done:
  grade = input()                          # read line from user
  if grade == '':                          # empty line was entered
    done = True
  elif grade not in points:                # unrecognized grade entered
    print("Unknown grade '{0}' being ignored".format(grade))
  else:
    num_courses += 1
    total_points += points[grade]
if num_courses > 0:                        # avoid division by zero
  print('Your GPA is {0:.3}'.format(total_points / num_courses))
```

```python
def add_number(a):
  a += 3  
  print(a)
  print(f'inside function: {id(a)}')

a = 0
print(a)
print(f'before function: {id(a)}')
add_number(a)
print(a)
print(f'after function: {id(a)}')
```

```python
def add_number(lst):
    lst.append(len(lst))
    print(lst)
    print("lst inside function after appending:" + str(id(lst)))


lst = [0]
print(lst)
print("lst outside function before execution:" + str(id(lst)))
add_number(lst)
print("lst outside function after execution:" + str(id(lst)))
print(lst)
```
